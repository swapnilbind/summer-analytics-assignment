### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Step 1: Reset index with explicit column naming
df = df.reset_index(names=['car_name'])  # Renames index during reset

# Step 2: Calculate median mpg
median_mpg = df['mpg'].median()

# Step 3: Filter cars with above-median mpg
above_median = df[df['mpg'] > median_mpg].copy()

# Step 4: Create horsepower-to-weight ratio
above_median['hp_weight_ratio'] = (
    above_median['horsepower'] / above_median['weight']
)

# Step 5: Find and display the best car
best_idx = above_median['hp_weight_ratio'].idxmax()
best_car = above_median.loc[best_idx, 'car_name']
best_car


```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group by year and origin to get average mpg
avg_mpg = df.groupby(['model_year', 'origin'], as_index=False)['mpg'].mean()

# Plotting the multi-line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Evolution of Average MPG Over the Years by Origin')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()

```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python

df = pd.read_csv('Cars.csv')
df.head()
car_counts = df.groupby('name')['model_year'].nunique().reset_index()
multi_year_cars = car_counts[car_counts['model_year'] > 1]['name']
consistency = (
    df[df['name'].isin(multi_year_cars)]
    .groupby('name')
    .agg(
        appearances=('mpg', 'count'),
        mpg_std=('mpg', 'std'),
        avg_mpg=('mpg', 'mean')
    )
    .reset_index()
)
consistent_cars = consistency[
    (consistency['appearances'] >= 2) & (consistency['mpg_std'] < 1.0)
]
if not consistent_cars.empty:
    min_std = consistent_cars['mpg_std'].min()
    best_cars = consistent_cars[consistent_cars['mpg_std'] == min_std]
else:
    best_cars = pd.DataFrame()
best_cars = best_cars.sort_values(
    by=['appearances', 'avg_mpg'], ascending=[False, False]
)
print(best_cars[['name', 'appearances', 'avg_mpg']])
```
